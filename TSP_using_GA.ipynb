```python
import random
import math
# Generate population size in each generation
population_size = 80
n = int(input())
```

```python
class Path:
    def __init__(self, string):
        self.string = string
        self.fitness = self.calculate_fitness()
    #Calculate fitness
    def calculate_fitness(self):
        fitness = 0
        for i in range(1, len(self.string)):
            distance = math.sqrt((dict[self.string[i]][0] - dict[self.string[i-1]][0])**2 + (dict[self.string[i]][1] - dict[self.string[i-1]][1])**2)
            fitness += distance
        return fitness
    #Check if whether the city in the path
    @classmethod
    def repeat(self, string, ch):
        if ch in string:
            return True
        return False
    @classmethod
    def random_genes(self):
        string = "0"
        while True:
            if len(string) == n:
                string += string[0]
                break

            temp = random.randint(1, n-1)
            if not self.repeat(string, chr(temp + 48)):
                string += chr(temp + 48)

        return string
    def mate(self, parent):
        child_string = []
        for child1, child2 in zip(self.string, parent.string):
            if not self.repeat(child_string, child1) and not self.repeat(child_string,child2):
                prob = random.random()
                if prob < 0.5:
                    child_string.append(child1)
                else:
                    child_string.append(child2)
            elif not self.repeat(child_string, child1):
                child_string.append(child1)
            elif not self.repeat(child_string,child2):
                child_string.append(child2)
            else:    
                for i in range(1, n):
                    if chr(i+48) not in child_string:
                        child_string.append(chr(i+48))
                        break
        child_string.append('0')
        return Path(child_string)                
```

```python
# Randomize the place of each city
dict = {chr(i+48): [random.randint(1,999), random.randint(1,999)] for i in range(n)}
generation = 1
iterations = 100
population = []
for i in range(population_size):
    random_path = Path(Path.random_genes())
    population.append(random_path)
while iterations > 0:
    iterations -= 1
    population = sorted(population, key=lambda x: x.fitness)
    print(f'Generation: {generation}  String: {"".join(population[0].string)}  Fitness: {population[0].fitness}')
    new_population = []
    new_population.extend(population[:10])
    for i in range(int(population_size*90/100)):
        par1 = random.choice(population[:50])
        par2 = random.choice(population[:50])
        child = par1.mate(par2)
        new_population.append(child)
    population = new_population
    generation += 1
```
